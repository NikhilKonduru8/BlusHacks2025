```python
import pandas as pd
import numpy as np

df = pd.read_excel("reviews.xlsx")

np.shape(df)[0]
```

```python
from keybert import KeyBERT

keyword = []

for i in range(np.shape(df)[0]):
  # Initialize the KeyBERT model
  model = KeyBERT('distilbert-base-nli-mean-tokens')

  # Example text
  text = df.Review[i]

  # Extract keywords
  keywords = model.extract_keywords(text)

  # Print the keywords

  keyword.append(keywords[0][0] + " " + keywords[1][0])
```

```python
keyword
```

```python
df_new = pd.DataFrame(keyword, columns=['Keywords'])
```
